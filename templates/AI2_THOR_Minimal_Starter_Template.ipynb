```
from ai2thor.controller import Controller
import matplotlib.pyplot as plt

print("正在切換到臥室場景 (FloorPlan304)...")

# 1. 重新啟動控制器，這次指定去臥室，並開啟深度圖
# 注意：Colab 上不需要指定版本，它會用剛剛裝好的最新版
controller = Controller(
    scene="FloorPlan304",      # 臥室場景 (有窗簾)
    gridSize=0.25,
    renderDepthImage=True,     # 【關鍵】開啟深度圖
    width=1000,
    height=600,
    server_start_timeout=300 # 增加伺服器啟動的超時時間，避免Colab環境下的TimeoutError
)

# 2. 移動機器人去看護窗簾
# 針對 FloorPlan304 的特定舞步：左轉兩次，前進兩次
print("正在移動機器人...")
controller.step(action="RotateLeft")
controller.step(action="RotateLeft")
controller.step(action="MoveAhead")
controller.step(action="MoveAhead")

# 3. 抓取畫面
event = controller.last_event
rgb = event.frame
depth = event.depth_frame

# 4. 畫圖出來 (這就是你要的截圖！)
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# 左邊：RGB
axes[0].imshow(rgb)
axes[0].set_title("RGB View (Soft Obstacle: Curtain)", fontsize=16)
axes[0].axis('off')

# 右邊：Depth
# 使用 'plasma' 配色，亮黃色代表近，深紫色代表遠
axes[1].imshow(depth, cmap='plasma')
axes[1].set_title("Depth View (Geometry info)", fontsize=16)
axes[1].axis('off')

plt.suptitle("Thesis Progress: VLM Soft Obstacle Navigation", fontsize=20)
plt.show()
```

```
import ai2thor.controller
import matplotlib.pyplot as plt

# --- 防呆機制：如果有舊的控制器卡住，先強制關閉 ---
try:
    if 'controller' in locals():
        controller.stop()
        print("已關閉舊的模擬器，釋放資源...")
except:
    pass
# ------------------------------------------------

print("正在啟動 AI2-THOR (這次給它 300秒，不會 Timeout)...")

# 1. 啟動控制器
# 我們加上 server_start_timeout=300，給它足夠的時間下載和啟動
controller = ai2thor.controller.Controller(
    scene="FloorPlan304",      # 臥室
    gridSize=0.25,
    renderDepthImage=True,     # 開啟深度圖
    width=1000,
    height=600,
    server_start_timeout=300   # 【關鍵修改】延長等待時間到 5 分鐘
)

print("成功啟動！正在前往窗簾位置...")

# 2. 移動機器人 (左轉x2, 前進x2)
controller.step(action="RotateLeft")
controller.step(action="RotateLeft")
controller.step(action="MoveAhead")
controller.step(action="MoveAhead")

# 3. 抓取畫面
event = controller.last_event
rgb = event.frame
depth = event.depth_frame

# 4. 畫圖
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# 左邊：RGB
axes[0].imshow(rgb)
axes[0].set_title("RGB View (Soft Obstacle)", fontsize=16)
axes[0].axis('off')

# 右邊：Depth
axes[1].imshow(depth, cmap='plasma')
axes[1].set_title("Depth View (Geometry)", fontsize=16)
axes[1].axis('off')

plt.suptitle("Proof of Concept: Soft Obstacle Navigation", fontsize=20)
plt.show()

print("搞定！圖片出來了嗎？")
```
